```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt  
import warnings 
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('oil_sales_assignment_dataset.csv')
```

```python
# Matplotlib
# Plot a line graph of total value_sales by month.
# Create a bar chart of total sales by city.
# Plot a histogram of average_price.
# Draw a pie chart showing the percentage of sales by class.
# Create a scatter plot of value_sales vs volume_sales.
# Plot a boxplot for average_price.
```

```python
# Plot a line graph of total value_sales by month. 
sales = df.groupby(['month', 'city'])['value_sales'].sum().reset_index()

sns.lineplot(data=sales, x='month', y='value_sales', hue='city', marker='o')
```

```python
sales = df.groupby(['month', 'manufacturer'])['value_sales'].sum().reset_index()

sns.lineplot(data=sales, x='month', y='value_sales', hue='manufacturer', marker='o')
```

```python
# Create a bar chart of total sales by city.
city_sale=df.groupby('city')['value_sales'].sum().reset_index()
sns.barplot(data=city_sale,y="city",x='value_sales',palette="rainbow")
plt.title("oil sales per city")
plt.xlabel('amount sold in $')
```

```python
# Plot a histogram of average_price.
sns.histplot(x=df['average_price'],kde=True)
```
